### 2. Are explicitly rated songs more popular than songs that are not explicit?

```python
explicit_songs = spotify[spotify.explicit == 1]
not_explicit_songs = spotify[spotify.explicit == 0]
e_var = explicit_songs.popularity.var()
ne_var = not_explicit_songs.popularity.var()
diff_between_e_and_ne = e_var - ne_var
```

```python
e_pop = explicit_songs.popularity.to_list()
ne_pop = not_explicit_songs.popularity.to_list()

# Perform Welch's t-test (for unequal variances)
t_statistic, p_value = stats.ttest_ind(e_pop, ne_pop, equal_var=False)
if t_statistic > 0:
    p_value /= 2

if (p_value < 0.05):
    print("One-tailed P-value T-test:", p_value)
    print("There is statistical significance\n")
else:
    print("Shit, no diff")
    
u_statistic, p_value_mwu = stats.mannwhitneyu(e_pop, ne_pop, alternative='greater')

if (p_value_mwu < 0.05):
    print("One-tailed P-value MWU:", p_value_mwu)
    print("There is statistical significance\n")
else:
    print("Shit, no diff")
```

### 3. Are songs in major key more popular than songs in minor key? 

```python
major = spotify[spotify['mode']== 1]
minor = spotify[spotify['mode']== 0]
major_var = major.popularity.var()
minor_var = minor.popularity.var()
diff_between_major_and_minor = major_var - minor_var
print("Major Var is " + str(major_var) + ",\n which differs " 
      + str(diff_between_major_and_minor) + " away from minor's var")
print("")

ma = major.popularity.to_list()
mi = minor.popularity.to_list()


t_statistic, p_value = stats.ttest_ind(ma, mi, equal_var=False)
if t_statistic > 0:
    p_value /= 2
else:
    p_value = 1 - p_value

if (p_value < 0.05):
    print(t_statistic)
    print("One-tailed P-value for T-test:", p_value)
    print("There is statistical significance")
else:
    print("One-tailed P-value for T-test:", p_value)
    print("No Statistical Significance")
    print("")

u_statistic, p_value_mwu = stats.mannwhitneyu(ma, mi, alternative='greater')
if (p_value_mwu < 0.05):
    print("One-tailed P-value for Man Whitney U:", p_value_mwu)
    print("There is statistical significance")
else:
    print("One-tailed P-value for Man Whitney U:", p_value_mwu)
    print("No Statistical Significance")
```

```python
def logistic_regression_analysis(X, y):
    # Median split
    X = np.array(X).reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=18147954)

    # Fit Logistic Regression model
    model = LogisticRegression()
    model.fit(X_train, y_train)

    # Predicted probabilities using cross-validation
    predicted_probs = cross_val_predict(model, X_train, y_train, method='predict_proba', cv=5)
    
    # Predictions for test set
    predicted_labels = model.predict(X_test)
    predicted_probs_test = model.predict_proba(X_test)[:, 1]

    # Calculate ROC AUC Score
    auc = roc_auc_score(y_test, predicted_probs_test)
    
    # Plot histogram of predicted probabilities
    plt.figure()
    plt.hist(predicted_probs[:, 1], bins=30, color='blue', alpha=0.7)
    plt.title('Predicted Probabilities Histogram')
    plt.xlabel('Predicted Probability')
    plt.ylabel('Frequency')
    plt.show()

    # Calculate F1 score, precision, and recall
    f1 = f1_score(y_test, predicted_labels)
    precision = precision_score(y_test, predicted_labels)
    recall = recall_score(y_test, predicted_labels)
    print("F1 Score:", f1)
    print("Precision:", precision)
    print("Recall:", recall)

    # Logistic Regression Curve
    X_range = np.linspace(X.min(), X.max(), 300)
    probabilities = model.predict_proba(X_range.reshape(-1, 1))[:, 1]

    plt.figure()
    plt.scatter(X, y, color='black', zorder=20)
    plt.plot(X_range, probabilities, color='red', linewidth=3)
    plt.title('Logistic Regression Curve')
    plt.xlabel('Average Movie Enjoyment')
    plt.ylabel('Probability of High Enjoyment')
    plt.show()

    return model.coef_[0][0], auc
coef__, auc_ = logistic_regression_analysis(X, y)
```
